```python
plt.plot(x,  y) 
plt.title("Squares") 
plt.xlabel("x-values") 
plt.ylabel("y-values") 
plt.xlim(-10,20)
plt.show()
```

```python

import numpy as np 
import pandas as pd 
np.random.seed(0) 
df = pd.DataFrame(data={'1':np.random.randint(0, 100, 10), '2':np.random.randint(0, 100, 10)})
plt.bar(df.index.values, df['1'])
plt.bar(df.index.values, df['2'], bottom = df['1'])
plt.legend(['1', '2']) 
plt.show()
```
