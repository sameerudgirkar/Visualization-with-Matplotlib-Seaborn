```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.shape
```

```python
iris.head()
```

```python
iris.species.unique()
```

# Univariate Analysis

Histogram -> Frequency distribution / For Univariate data

```python
iris['petal_length']
```

```python
iris['petal_length'].value_counts().head(5)
```

```python
sns.histplot(data = iris, x = 'petal_length', bins = 50)
plt.show()
```

```python
# Bar plot -> requires Categorical data
# Scatter plot -> requires 2 variables
# Box plot -> requires Univariate Continuos/Numerical variables
```

# KDE (kernel density distribution)

Smooth version of histogram

```python
sns.kdeplot(data = iris,x = 'petal_length')
plt.show()
```

Histogram -> Count

KDE       -> Density (Observations/Total Count)

```python
sns.kdeplot(data = iris,x = 'petal_length', hue = 'species')
plt.show()
```

# Distribution Plot

```python
sns.displot(data = iris, x = 'sepal_length', bins = 40, hue = 'species')
plt.show()
```

# Bivariate Analysis

- 2 Continuos/Numerical variables

```python
sns.scatterplot(data = iris, x = 'petal_length', y = 'petal_width',hue = 'species')
# plt.xlabel("My X-axis")
# plt.ylabel("My Y-axis")
# plt.xlim(-5,10)
plt.show()
```

**Observations**:
Same plot using matplotlib plot--- Increased lines of code as compared to matplotlib's library called seaborn(sns)

```python
setosa = iris[iris['species'] == 'setosa']
versicolor = iris[iris['species'] == 'versicolor']
virginica = iris[iris['species'] == 'virginica']
```

```python
plt.scatter(x = setosa['petal_length'],y = setosa['petal_width'], c = 'blue')
plt.scatter(x = versicolor['petal_length'],y = versicolor['petal_width'], c = 'orange')
plt.scatter(x = virginica['petal_length'],y = virginica['petal_width'], c = 'green')
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.show()
```

```python
sns.jointplot(data=iris,x = 'sepal_length', y = 'sepal_width',kind = 'hex',color = 'blue')
plt.show()
```
